```r
#
# Example of multiple linear regression
#
# Read data from a text file
data2<-read.table("table5.1.txt",header=TRUE)
print(data2)
```

```r
# plot samples
plot(data2$x1,data2$y)
text(data2$x1, data2$y, labels = seq_along(data2$x1), pos=4, col= 'red')
```

```r
plot(data2$x2,data2$y)
text(data2$x2, data2$y, labels = seq_along(data2$x2), pos=4, col= 'red')
```

```r
# linear regression
x <- data2[,1:2]
y <- data2$y

reg <- lm(y~.,data=x)
summary(reg)
```

```r
plot(reg)
```
