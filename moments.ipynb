指数分布のモーメントの計算

```r
options(repr.plot.width=5, repr.plot.height=5)
```

指数分布に従う乱数の生成

```r
lambda = 5
x = rexp(10000, lambda)
hist(x, freq=FALSE)
```

モーメント計算のためのパッケージの読み込み

```r
library(e1071)
```

１次モーメント

```r
moment(x, order=1, center=FALSE) 
print(1/lambda)
```

２次モーメント

［注］n! = fracorial(n) または n! = gamma(n+1)

```r
moment(x, order=2, center=FALSE) 
print(factorial(2)/(lambda*lambda))
```

平均周りの２次モーメント

```r
moment(x, order=2, center=TRUE) 
print(1/(lambda*lambda))
```

3次モーメント

```r
moment(x, order=3, center=FALSE) 
print(factorial(3)/(lambda**3))
```

平均周りの3次モーメント

```r
moment(x, order=3, center=TRUE) 
```

4次モーメント

```r
moment(x, order=4, center=FALSE) 
print(factorial(4)/(lambda**4))
```

平均周りの4次モーメント

```r
moment(x, order=4, center=TRUE) 
```
