### 中央極限定理のデモ（一様分布の場合）

```r
options(repr.plot.width=5, repr.plot.height=5)
```

```r
library(MASS)

x = runif(100000)
hist(x, breaks=seq(0,1,0.05), freq=FALSE)
```

２つの一様分布の和

```r
x = runif(100000) + runif(100000)
x = x / 2
hist(x, breaks=seq(0,1,0.05), freq=FALSE)
(x.normfit <- fitdistr(x,"normal"))
curve(dnorm(x,mean=x.normfit$estimate[1],sd=x.normfit$estimate[2]),col="blue",add=TRUE)
```

４つの一様分布の和

```r
x = runif(100000) + runif(100000) + runif(100000) + runif(100000)
x = x / 4
hist(x, freq=FALSE, breaks=seq(0,1,0.05))
(x.normfit <- fitdistr(x,"normal"))
curve(dnorm(x,mean=x.normfit$estimate[1],sd=x.normfit$estimate[2]),col="blue",add=TRUE)
```

12個の一様分布の和

```r
x = runif(100000) + runif(100000) + runif(100000) + runif(100000) + runif(100000) + runif(100000) + runif(100000) + runif(100000) + runif(100000) + runif(100000) + runif(100000) + runif(100000)
x = x / 12
hist(x, breaks=seq(0,1,0.05), freq=FALSE)
(x.normfit <- fitdistr(x,"normal"))
curve(dnorm(x,mean=x.normfit$estimate[1],sd=x.normfit$estimate[2]),col="blue",add=TRUE)
```
