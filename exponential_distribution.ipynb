Plot of Exponential Distribution

```r
options(repr.plot.width=5, repr.plot.height=5)
```

```r
x <- seq(0, 5, length=100)
```

```r
lambda <- 1.0
hx <- dexp(x, rate=lambda)
title <- sprintf("Exponential Distributions lambda=%2.1f", lambda)
plot(x, hx, type="l", lty=1, xlab="x value", ylab="Density", main=title)
```

```r
lambda <- 3.0
hx <- dexp(x, rate=lambda)
title <- sprintf("Exponential Distributions lambda=%2.1f", lambda)
plot(x, hx, type="l", lty=1, xlab="x value", ylab="Density", main=title)
```

```r
lambda <- 0.5
hx <- dexp(x, rate=lambda)
title <- sprintf("Exponential Distributions lambda=%2.1f", lambda)
plot(x, hx, type="l", lty=1, xlab="x value", ylab="Density", main=title)
```

### curve

```r
lambda=1.0
curve(dexp(x, rate=lambda), 0, 5)
```
