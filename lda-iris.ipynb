# 　アヤメのデータの線形判別分析

アヤメのデータの線形判別分析をRで実行してみます．

アヤメのデータは，Rにすでに登録されています．
それを読み込むためにライブラリを読み込みます．

```r
library(MASS)
```

アヤメのデータは，irisという変数に読み込まれています．まずは，読み込んだアヤメのデータを見てみます．

```r
iris
```

散布図を描いてみます．

```r
plot(iris[,1:4], col=iris$Species )
```

150個のデータのうちの７割のデータを訓練用データとして判別分析で判別写像を構成し，残りの３割のデータで構成した判別写像を使った識別手法の性能を評価してみます．
まずは，７割のデータで判別分析を行い，訓練用のサンプルを２次元の判別空間にプロットしてみます．

```r
# Set Training Sample Index
train.index <- sample(nrow(iris), nrow(iris)*0.7)

# Fisher's Linear Discriminant Analysis for Training Samples
res.lda <- lda(Species~., iris, subset=train.index, method="moment")

# plot of Training Samples in Discriminant Space
#plot(res.lda)
plot(predict(res.lda,iris)$x, pch = as.integer(iris$Species), col = iris$Species)
```

最後に，訓練データとテスト用のデータのそれぞれに対する識別性能（認識率）を計算して表示してみます．

```r
# Recognition Rate for Training Samples
recogtrain <- 1.0 - mean(predict(res.lda, iris[train.index,])$class != iris[train.index,]$Species)

print('Recognition Rate (Training)')
print(recogtrain)

# Recognition Rate for Test Samples
recogtest <- 1.0 - mean(predict(res.lda, iris[-train.index,])$class != iris[-train.index,]$Species)

print('Recognition Rate (Test)')
print(recogtest)
```
