## 2D Gaussioan Distribution

```r
options(repr.plot.width=5, repr.plot.height=5)
```

```r
install.packages('mvtnorm')
```

平均　$\mu = \begin{bmatrix} 0 \\ 0 \end{bmatrix}$，分散共分散行列　$\Sigma = \begin{bmatrix} 2 & 0.8 \\ 0.8 & 1 \end{bmatrix}$の２次元正規分布

```r
library(mvtnorm)

sigma = matrix(c(2, 0.8, 0.8, 1), ncol = 2)

x1 = seq(-5, 5, length = 50)
x2 = x1
f = function(x1, x2) {
  dmvnorm(matrix(c(x1, x2), ncol = 2), mean = c(0, 0), sigma = sigma)
}
z = outer(x1, x2, f)
z[is.na(z)] = 1
op = par(bg = "white")
persp(x1, x2, z, theta = 30, phi = 30, expand = 0.5, col = "orange")
```

２次元正規分布からのサンプル（乱数）の２次元プロット

```r
rand = rmvnorm(n = 1000, mean = c(0, 0), sigma)
x1 = rand[, 1]
x2 = rand[, 2]
plot(x1, x2)
```

正規化の例の2次元正規分布

```r
mean = c(1,2)
sigma = matrix(c(5, 4, 4, 5), ncol = 2)

x1 = seq(-3, 6, length = 50)
x2 = x1
f = function(x1, x2) {
  dmvnorm(matrix(c(x1, x2), ncol = 2), mean = mean, sigma = sigma)
}
z = outer(x1, x2, f)
z[is.na(z)] = 1
op = par(bg = "white")
persp(x1, x2, z, theta = 30, phi = 30, expand = 0.5, col = "orange")
```

```r
rand = rmvnorm(n = 1000, mean = mean, sigma)
x1 = rand[, 1]
x2 = rand[, 2]
plot(x1, x2)
```

平均を$0$，分散共分散行列を単位行列に正規化した空間の正規分布

```r
mean = c(0,0)
sigma = matrix(c(1, 0, 0, 1), ncol = 2)

x1 = seq(-4, 4, length = 50)
x2 = x1
f = function(x1, x2) {
  dmvnorm(matrix(c(x1, x2), ncol = 2), mean = mean, sigma = sigma)
}
z = outer(x1, x2, f)
z[is.na(z)] = 1
op = par(bg = "white")
persp(x1, x2, z, theta = 30, phi = 30, expand = 0.5, col = "orange")
```

```r
rand = rmvnorm(n = 1000, mean = mean, sigma)
x1 = rand[, 1]
x2 = rand[, 2]
plot(x1, x2)
```
