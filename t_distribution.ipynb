### Plot t-distribution

```r
options(repr.plot.width=5, repr.plot.height=5)
```

```r
x <- seq(-5, 5, length=100)
```

```r
dfree <- 5-1
hx <- dt(x, df=dfree)
title <- sprintf("t Distribution t(%d)", dfree)
plot(x, hx, type="l", lty=1, xlab="x value",
  ylab="Density", main=title)
```

```r
dfree <- 10-1
title <- sprintf("t Distribution t(%d)", dfree)
plot(x, hx, type="l", lty=1, xlab="x value",
  ylab="Density", main=title)
```
