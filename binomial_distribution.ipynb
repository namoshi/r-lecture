Plot of Binomial Distribution

```r
options(repr.plot.width=5, repr.plot.height=5)
```

```r
x <- seq(0, 12, length=13)
print(x)
```

```r
p=0.5
hx <- dbinom(x, size=12, prob=p)
hx
title <- sprintf("Binomial Distributions (p=%f)", p)
plot(x, hx, lty=2, xlab="x value", ylab="Density", main=title)
```
