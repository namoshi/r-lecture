### Plot of chi-suare Distribution

```r
options(repr.plot.width=5, repr.plot.height=5)
```

```r
x <- seq(0, 50, length=100)
```

```r
dfree <- 3-1
hx <- dchisq(x, df=dfree)
title <- sprintf("Chi-Square Distribution dfree = %d", dfree)
plot(x, hx, type="l", lty=1, xlab="x value", ylab="Density", main=title)
```

```r
dfree <- 5-1
hx <- dchisq(x, df=dfree)
title <- sprintf("Chi-Square Distribution dfree = %d", dfree)
plot(x, hx, type="l", lty=1, xlab="x value", ylab="Density", main=title)
```
