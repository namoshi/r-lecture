Plot of 1-D Gaussian Distribution

```r
options(repr.plot.width=5, repr.plot.height=5)
```

```r
x <- seq(-5, 5, length=100)
x
```

```r
mu <- 0.0
sigma <- 1.0
hx <- dnorm(x, mean=mu, sd=sigma)
title <- sprintf("Gaussian Distributions N(%2.1f, %2.1f)", mu, sigma)
plot(x, hx, type="l", lty=1, xlab="x value", ylab="Density", main=title)
```
