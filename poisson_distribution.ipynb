Plot of Poisson Distribution

```r
options(repr.plot.width=5, repr.plot.height=5)
```

```r
x <- seq(0, 12, length=13)
x
```

```r
l=1
hx <- dpois(x, lambda=l)
hx
title <- sprintf("Poisson Distributions (Lambda=%f)", l)
plot(x, hx, lty=2, xlab="x value", ylab="Density", main=title)
```

```r
l=3
hx <- dpois(x, lambda=l)
hx
title <- sprintf("Poisson Distributions (Lambda=%f)", l)
plot(x, hx, lty=2, xlab="x value", ylab="Density", main=title)
```
