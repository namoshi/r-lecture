# 成績データの主成分分析¶

成績データファイルを作成し，読み込んでください．

以下の例では，データファイルをテキストエディタで作成して，それをRに読み込んでいますが，前回のようにエクセルでcsv形式のファイルを作成して，それを read.csvで読み込んでも構いません．

```r
# Reade Seiseki Data from data file
#
seisekidata = read.table("table9.1.txt",header=TRUE)
print(seisekidata)
```

成績データのプロット

```r
plot(seisekidata)
```

主成分分析の実行にはprincompを使います．
summaryで結果の要約を表示しています．

```r
#
# Apply PCA to the seiseki data
#
pca.seiseki <- prcomp(seisekidata)
print(pca.seiseki)
#
# Print Summary of the PCA
#
summary(pca.seiseki)
```

主成分分析の結果を第１主成分と第２主成分の空間にプロットします．
これにより，４次元のデータを２次元で見ることができるようになります．

```r
#
# Plot the results of PCA (The first and the second principal components)
#
plot(pca.seiseki$x[,1], pca.seiseki$x[,2])
text(pca.seiseki$x[,1], pca.seiseki$x[,2], 
     labels = seq_along(seisekidata[,1]), pos=4, col= 'red')
```
